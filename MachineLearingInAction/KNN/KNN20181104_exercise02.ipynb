#### 1、将每一个32* 32的二进制图像矩阵都转化成1* 1024的向量

```python
def img2vector(filename):
    returnVect=np.zeros((1,1024))
    fr=open(filename)
    for i in range(32):
        lineStr=fr.readline()
        for j in range(32):
            returnVect[0,32*i+j]=int(lineStr[j])
    return returnVect
```

#### 2、测试算法

```python
from os import listdir
def handwritingClassTest():
    hwLabels=[] # 创建空向量，作为训练集标签向量
    trainingFileList=listdir('E:/pythonStudy/MachineLearingInAction/machinelearninginaction/Ch02/trainingDigits')  # 获取训练集文件目录
    m=len(trainingFileList)
    traingMat=np.zeros((m,1024)) # 创建训练集矩阵
    for i in range(m):  # 对于每个训练数据，做如下处理
        fileNameStr=trainingFileList[i]
        fileStr=fileNameStr.split('_')[0]
        classNumStr=int(fileStr.split('.')[0])  # 获得每个文件的类标签（即文件名的开始数字）
        hwLabels.append(classNumStr)  # 将类标签添加到hwLabels中
        # 将每个处理完的训练数据添加到训练样本
        traingMat[i,:]=img2vector('E:/pythonStudy/MachineLearingInAction/machinelearninginaction/Ch02/trainingDigits/%s' % fileNameStr)  
    
    # 测试样本
    testFileList=listdir('E:/pythonStudy/MachineLearingInAction/machinelearninginaction/Ch02/trainingDigits')  
    errorCount=0.0
    mTest=len(testFileList)
    for i in range(mTest):
        fileNameStr=testFileList[i]
        fileStr=fileNameStr.split('.')[0]
        classNumStr=int(fileStr.split('_')[0])
        vectorUnderTest=img2vector('E:/pythonStudy/MachineLearingInAction/machinelearninginaction/Ch02/trainingDigits/%s' % fileNameStr)
        # 使用KNN算法模型进行训练
        classifilerResult=KNN.classify0(vectorUnderTest,traingMat,hwLabels,3)
        print ("the classifier came back with: %d, the real answer is: %d" % (classifilerResult,classNumStr))
        if(classifilerResult != classNumStr) : errorCount +=1.0
    print ("\nthe total number of errors is: %d" % errorCount)
    print ("\nthe total error rate is: %d" % (errorCount/float(mTest)))
```

```python
handwritingClassTest()
```
